```python
# Importing all the necessary libraries

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from selenium.webdriver.common.action_chains import ActionChains
from webdriver_manager.chrome import ChromeDriverManager

from time import sleep
from csv import writer
from re import match
from requests import ConnectionError
```

```python
def browseSite1(hadithCounter, writeMaster, writeNarrator, writeReference, writeHCRB):
    try:
        # Saving the url in url variable
        url = "https://www.ihsanetwork.org/hadith.aspx"
        
        # Selecting Chrome as our browser and installing Chrome Drivers 
        browser = webdriver.Chrome(ChromeDriverManager().install())

        # Opening the site in chrome
        browser.get(url)

        # Making chain action on browser
        action = ActionChains(browser)

        # Finding صحيح مسلم button
        صحيح_مسلمBtn = browser.find_element(By.XPATH, "//span[@id='dijit__TreeNode_2_label']")

        sleep(1)

        # First right clicking on صحيح مسلم button
        action.context_click(صحيح_مسلمBtn).perform()

        # Now finding عرض النص button which will land us to hadith page
        ##عرض_النصBtn = browser.find_element(By.XPATH, "//td[@id='dijit_MenuItem_1_text']")

        # Writing a hadith number in text box and hitting enter
        عرض_حديث = browser.find_element(By.XPATH, "/html[1]/body[1]/div[2]/table[1]/tbody[1]/tr[4]/td[3]/div[1]/div[1]/input[1]")

        عرض_حديث.send_keys(str(hadithCounter), Keys.ENTER)

        sleep(1)

        # Now clicking عرض_النص button which will land me to hadith number one page
        ##action.click(عرض_النصBtn).perform()

        extratDataFromSite(hadithCounter, writeMaster, writeNarrator, writeReference, writeHCRB, browser)
    except ConnectionError:
        print("\nAgain Reaching Site for hadith: ", hadithCounter, end="\n")
        currentHadith = hadithCounter
        print("Current Hadith is: ", currentHadith, end="\n\n")
        sleep(1)
        browseSite(currentHadith, writeMaster, writeNarrator, writeReference, writeHCRB)
    except:
        print("\nAgain Reaching Site for hadith: ", hadithCounter, end="\n")
        currentHadith = hadithCounter
        print("Current Hadith is: ", currentHadith, end="\n\n")
        sleep(1)
        browseSite(currentHadith, writeMaster, writeNarrator, writeReference, writeHCRB)
```

```python
def extratDataFromSite(hadithCounter, writeMaster, writeNarrator, writeReference, writeHCRB, browser):
    try:
        while True:
            #sleep(5)

            # Finding form
            form = browser.find_elements(By.XPATH, "//form[@id='Form1']")

            sleep(20)

            resultArray = []

            # Saving form data in array
            for result in form:
                resultArray = result.text.splitlines()

            # Printing the hole array
            ##print(resultArray, end="\n\n")
            # Only printing the title of the current page
            ##print(resultArray[4], end="\n\n")
            # Only printing the listner of hadith
            ##print(resultArray[8::],  end="\n\n")

            # Spliting the title so we can separate the book name, chapter and topic from each other
            title = resultArray[4].split('،')

            # Now spliting حديث and صحيح مسلم from each other
            book = title[0].split('حديث')

            text = ""

            for i in book:
                if i != '':
                    text = i

            title[0] = text

            # Finding التالى button which will land us to next hadith/page
            التالىBtn = browser.find_element(By.XPATH, "//a[contains(text(),'التالى')]")

            finalTitle = []

            sannutCounter = 0
            orderCounter = 0
            listner = []
            narratorList = []

            takhrij = []
            finalTakhrij = []
            lastDate = ''

            finalHCRB = []
            refBook = []

            # Checking if title start with صحيح مسلم then skip that page otherwise process the hole page and gather the required info
            # and write it in excel file
            if title[0] != 'صحيح مسلم':
                print('Manual hadith number: ', hadithCounter, end="\n\n")

                # Adding hadith number in list
                finalTitle.append(str(hadithCounter))

                # Reversing the title so we can write the master file in right order 
                title.reverse()

                # Also adding book name, topic and chapter in list
                for t in title:
                    finalTitle.append(t)

                print(finalTitle, end="\n\n")

                # Writing a row in master file
                writeMaster.writerow(finalTitle)

                # We will append every listner in listner list untill we receive *************** and when this pattern is detected
                # then we will increament in sannutCounter after that we will reverse the list of listner so that we could have the
                # right order of sunnat and in end we will clear the listner list
                for listen in resultArray[8::]:
                    orderCounter = 0
                    if listen == '***************':
                        sannutCounter += 1
                        listner.reverse()
                        print("Sunnat number: ", sannutCounter, end='\n\n')
                        print(listner, end='\n\n')

                        for l in listner:
                            orderCounter += 1
                            narratorList.extend([str(hadithCounter), str(sannutCounter), str(orderCounter), l])
                            writeNarrator.writerow(narratorList)
                            narratorList.clear()

                        listner.clear()
                    else:
                        listner.append(listen)

                # The last listner were added to the listner list but cannot be place in there original order due the end of loop
                # therefore we have done this code so that all the listner have there original order
                orderCounter = 0
                sannutCounter += 1
                listner.reverse()
                print("Sunnat number: ", sannutCounter, end='\n\n')
                print(listner, end='\n\n')

                # Writing narrator file
                for l in listner:
                    orderCounter += 1
                    narratorList.extend([str(hadithCounter), str(sannutCounter), str(orderCounter), l])
                    writeNarrator.writerow(narratorList)
                    narratorList.clear()

                print("التخريج", end="\n\n")

                # Finding التخريج button
                التخريجBtn = browser.find_element(By.XPATH, "//span[contains(text(),'التخريج')]")

                # Clicking on التخريج button
                التخريجBtn.click()

                # Finding التخريج div
                context = browser.find_element(By.XPATH, "/html[1]/body[1]/div[1]/form[1]/div[3]/div[3]/div[3]/div[1]/div[5]/div[1]/div[3]/div[2]/div[1]")

                # Spliting التخريج div text on the basis of \n
                context = context.text.split('\n')

                content = []

                # Adding all the text of التخريج div except '' and كتاب التخريج رقم الربط الموطن 
                for text in context:
                    if text != '':
                        if text != 'كتاب التخريج رقم الربط الموطن':
                            content.append(text)

                # Saving the التخريج first date in lastDate variable
                lastDate = content[1]

                date = ''

                refHadith = ''

                # Now reading التخريج data from index 2 till end and appending those indexes in list named as takhrij and when the string
                # of التخريج end with 'ل' then spliting the takhrij list index values one by one then making a string of those values
                # then making a final list named as finalTakrij which will help us to write data of التخريج in reference file but first
                # we have to reverse the list so that we can restore the original/required order
                for t in content[2::]:
                    if t.endswith('ل'):
                        date = lastDate
                        lastDate = t

                        ref = []

                        for r in takhrij:
                            ref = r.split('-')

                            ##print('ref: ', ref, end='\n\n')

                            # Saving chapter and hadith number in chapter variable
                            chapter = ref[len(ref)-1]

                            # Spliting chapter and hadith number
                            finalChapter = chapter.split('حديث')

                            # Matching the last index number with RE (Regular Expression)
                            if match(u'[\u0660-\u0669]+$', finalChapter[len(finalChapter)-1]):
                                refHadith = finalChapter[len(finalChapter)-1]
                                finalChapter.remove(refHadith)

                            # Making the chapter name  
                            chapterString = ''

                            for c in finalChapter:
                                chapterString += c

                            # Updating the last index of ref with only chapter name
                            ref[len(ref)-1] = chapterString

                            # Adding reference hadith number in ref
                            ref.append(refHadith)

                            ##print('updated ref: ', ref, end='\n\n')

                            string = ''

                            for fr in ref:
                                if fr != "":
                                    string = fr+","+string

                            ##print('string: ', string, end='\n\n')

                            finalRef = string.split(',')

                            ##print('final ref: ', finalRefHadith, end='\n\n')

                            refBook.append(finalRef[3])

                            sleep(1)

                            finalTakhrij.extend([str(hadithCounter), content[0], date, "|", finalRef[0], finalRef[1], finalRef[2], finalRef[3]])
                            finalTakhrij.reverse()

                            print(finalTakhrij, end='\n\n')

                            writeReference.writerow(finalTakhrij)

                            finalTakhrij.clear()
                        takhrij.clear()
                    else:
                        takhrij.append(t)
                        date = lastDate

                # The last التخريج data was added in takhrij list but cannot be place in its required order due the end of
                # loop therefore we have done this code so that all the التخريج data have its required order
                for r in takhrij:
                    ref = r.split('-')

                    ##print('ref: ', ref, end='\n\n')

                    # Saving chapter and hadith number in chapter variable
                    chapter = ref[len(ref)-1]

                    # Spliting chapter and hadith number
                    finalChapter = chapter.split('حديث')

                    # Matching the last index number with RE (Regular Expression)
                    if match(u'[\u0660-\u0669]+$', finalChapter[len(finalChapter)-1]):
                        refHadith = finalChapter[len(finalChapter)-1]
                        finalChapter.remove(refHadith)

                    # Making the chapter name  
                    chapterString = ''

                    for c in finalChapter:
                        chapterString += c

                    # Updating the last index of ref with only chapter name
                    ref[len(ref)-1] = chapterString

                    # Adding reference hadith number in ref
                    ref.append(refHadith)

                    ##print('updated ref: ', ref, end='\n\n')

                    string = ''

                    for fr in ref:
                        if fr != "":
                            string = fr+","+string

                    ##print('string: ', string, end='\n\n')

                    finalRef = string.split(',')

                    ##print('final ref: ', finalRefHadith, end='\n\n')

                    refBook.append(finalRef[3])

                    sleep(1)

                    finalTakhrij.extend([str(hadithCounter), content[0], date, "|", finalRef[0], finalRef[1], finalRef[2], finalRef[3]])
                    finalTakhrij.reverse()

                    print(finalTakhrij, end='\n\n')

                    writeReference.writerow(finalTakhrij)

                    finalTakhrij.clear()
                    takhrij.clear()

                ##print(content, end="\n\n")

                ##print('ref Books: ', refBook, end="\n\n")

                checkBook = []

                checkBook.append(refBook[0])

                # Saving the ref book name one time in checkBook list so we can create a finite RE with the help of that list
                for rB in refBook:
                    ##print('Ref Book', rB, end='\n\n')
                    flag = False

                    for cB in checkBook:
                        # Checking if the ref book name already exist in checkBook list or not if it exist then save True in flag
                        if rB == cB:
                            flag = True

                    # Here we are check the value of flag if its value is not True then add ref Book in checkBook list therefore
                    # all the ref Book will be reported one time
                    if flag != True:
                        checkBook.append(rB)

                ##print('Check Book', checkBook, end='\n\n')

                for cB in checkBook:
                    refHadithCounter = 0
                    for rB in refBook:
                        # Matching the checkBook value with ref Book value if it match then increament in refHadithCounter variable  
                        if match(cB, rB):
                            refHadithCounter += 1

                    print('Hadith Count: ', refHadithCounter, ' in Book: ', cB, end='\n\n')

                    sleep(1)

                    finalHCRB.extend([str(hadithCounter), str(refHadithCounter), cB])

                    writeHCRB.writerow(finalHCRB)

                    finalHCRB.clear()

                sleep(1)

                # Click on التالى button
                التالىBtn.click()

                if hadithCounter == 2500:
                    break

                hadithCounter += 1
                
                sleep(1)
            else:
                # Click on التالى button
                التالىBtn.click()
    except ConnectionError:
        currentHadith = hadithCounter
        print("Current Hadith is: ", currentHadith, end="\n\n")
        sleep(1)
        browseSite2(currentHadith, writeMaster, writeNarrator, writeReference, writeHCRB, browser)
    except:
        currentHadith = hadithCounter
        print("Current Hadith is: ", currentHadith, end="\n\n")
        sleep(1)
        browseSite2(currentHadith+1, writeMaster, writeNarrator, writeReference, writeHCRB, browser)
```

```python
def removeDuplication(in_file, out_file):
    seen = []
    flag = False
    
    for line in in_file:
        for duplicate in seen:
            if line == duplicate:
                flag = True
                break
            else:
                flag = False

        if flag != True:
            seen.append(line)
            out_file.write(line)
```
